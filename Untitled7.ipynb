```python
from gramformer import Gramformer
import gradio as gr

# Instantiate Gramformer
gf = Gramformer(models = 1, use_gpu=False)

def correct(sentence):
    res = gf.correct(sentence)
    return res[0]

take_input = gr.inputs.Textbox(lines=3, placeholder='Enter an incorrect sentence here')

interFace = gr.Interface(
    fn=correct,
    inputs=take_input,  # Correct the typo here
    outputs='text',
    title="Hi there, I'm Gramformer"
)

interFace.launch()  # Correct the typo here
```
