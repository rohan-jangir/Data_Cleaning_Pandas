```python
job_salary = df.groupby('job_category')['salary']
salary = df['salary']
job_cat = df['job_category']
plt.xlabel('Jobs')
plt.ylabel('Salary')
plt.title('Salary Analysis Basis on Jobs')
plt.xticks(rotation=70)
plt.bar(job_cat,salary)
plt.show()
plt.savefig('salary basis on jobs.png')
```

```python
emp_resis = df['employee_residence']
salary = df['salary_in_usd']
plt.subplots(figsize=(20,7))
plt.xlabel('Country',fontsize=30)
plt.ylabel("Salary(usd)",fontsize=30)
plt.title("Salary(usd) Analysis Basis on Employee Resisdence")
plt.xticks(rotation=90,fontsize=13)
plt.bar(emp_resis,salary)
plt.show()
```

```python
ex_level = df['experience_level']
salary_usd = df['salary_in_usd']
plt.xlabel('Experience',fontsize=20)
plt.ylabel("Salary(usd)",fontsize=20)
plt.title("Salary(usd) Analysis Basis on Employee Experience")
plt.bar(ex_level,salary_usd)
plt.show()
```

```python
comp_size = df['company_size']
salary_usd = df['salary_in_usd']
plt.xlabel('Company Size',fontsize=20)
plt.ylabel("Salary(usd)",fontsize=20)
plt.title("Salary(usd) Analysis Basis on Company Size")
plt.bar(comp_size,salary_usd)
plt.show()
```
