```python
sale_tree = df.groupby('Year')['Number of trees sold'].sum()
plt.figure(figsize=(10,8))
plt.pie(sale_tree,labels=sale_tree,autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('sale of no. of trees')
plt.show()
```
